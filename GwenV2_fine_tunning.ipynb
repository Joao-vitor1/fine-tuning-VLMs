```python
system_message = """You are a highly advanced Vision Language Model (VLM), specialized in analyzing, describing, and interpreting visual data.
Your task is to process and extract meaningful insights from images, videos, and visual patterns,
leveraging multimodal understanding to provide accurate and contextually relevant information."""

def format_data(dados):
  image_name = dados['image_id'] + '.png'
  image = Image.open(os.path.join(path_imagens, image_name))
  return [
      {"role":"system", "content":[{'type':"text", "text": system_message}]},
      {"role":'user', 'content':[{'type':'image', 'image': image},
                                 {'type':'text', 'text':dados['question']}]},
      {'role':'assistant', 'content':[{'type':'text', 'text':dados['answer']}]}
  ]
```

```python
df_train = data_train.apply(format_data, axis=1).to_list()
df_eval = data_eval.apply(format_data, axis=1).to_list()
```

```python
import os
os.environ['WANDB_DISABLED'] = "true"

import torch
from transformers import Qwen2VLForConditionalGeneration, Qwen2VLProcessor, BitsAndBytesConfig
from peft import LoraConfig, get_peft_model
from trl import SFTConfig, SFTTrainer

import warnings
warnings.filterwarnings('ignore')
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

MODEL_ID = "Qwen/Qwen2-VL-7B-Instruct"
```

```python
if device == "cuda":
    bnb_config = BitsAndBytesConfig(
        load_in_4bit=True,
        bnb_4bit_use_double_quant=True,
        bnb_4bit_quant_type="nf4",
        bnb_4bit_compute_dtype=torch.bfloat16
    )
    model = Qwen2VLForConditionalGeneration.from_pretrained(
      MODEL_ID,
      device_map="auto",
      torch_dtype=torch.bfloat16,
      quantization_config=bnb_config,
      use_cache=False
  )
else:
    model = Qwen2VLForConditionalGeneration.from_pretrained(
        MODEL_ID,
        device_map="auto",
        torch_dtype=torch.bfloat16,
        use_cache=False
    )
processor = Qwen2VLProcessor.from_pretrained(MODEL_ID)
processor.tokenizer.padding_side = "right"
```

```python
def text_generator(exemplo):
  text = processor.apply_chat_template(
      exemplo[:2], #não pegar resposta
      tokenizer = False,
      add_generation_prompt = True # adiciona <|im_start|>assistant ao final
  )
  # print(f"Prompt: {text}")
  # print("-"*30)


  image_inputs = exemplo[1]["content"][0]["image"]
  input = processor(text = [text], images = image_inputs, return_tensors = 'pt').to(device)
  # print(input)
  # print("-"*30)


  generated_ids = model.generate(**input,
                                 max_new_tokens = 1024 #máximo de tokens gerados na resposta
                                 ).to(device)

  resposta_modelo = processor.decode(generated_ids[0],
                        skip_special_tokens=True, #esconder tokens especiais
                        clean_up_tokenization_spaces=False #???
                        )
  del input
  resposta_correta = exemplo[2]['content'][0]['text']

  return resposta_modelo, resposta_correta
```

```python
peft_config = LoraConfig(
    lora_alpha = 16, #geralmente r*2, alto = rapido aprendizado/ baixo = regularização
    lora_dropout = 0.1, # implementa regularização
    r = 8, #almenta o número de parâmetros
    bias = "none",
    target_modules = ['q_proj','v_proj'],
    task_type = 'CAUSAL_LM' #tipo de tarefa (CAUSAL_LM = prever próximo token)
)
```

```python
training_args = SFTConfig(
    output_dir= "gwen_tunned",
    num_train_epochs=2, #número de epocas
    per_device_train_batch_size=1, #número de batchs de treino
    per_device_eval_batch_size=1, #número de batchs de validação
    gradient_checkpointing=True,
    learning_rate=2e-5, #learning rate
    eval_strategy='steps', #calcular loss por step
    save_strategy='steps', #salvar modelo a cada taxa determinada de step
    logging_steps=5, #calcular loss a cada 5 steps
    eval_steps=5,  #calcular loss de validação a cada 5 steps
    save_steps=5,  #taxa de steps para salvar modelo
    metric_for_best_model='eval_loss', #metrica para definir melhor modelo
    load_best_model_at_end=True, #
    max_grad_norm=1, #
    warmup_steps=0, #
    dataset_kwargs={"skip_prepare_dataset": True}, # We have to put for VLMs
    max_seq_length=128, #
    remove_unused_columns= False, # VLM thing
    optim= "paged_adamw_32bit", #otimizador
)
```

```python
def collate_fn(exemplos):
  textos = [processor.apply_chat_template(exemplo, tokenizer=False) for exemplo in exemplos]
  image_inputs = [exemplo[1]['content'][0]['image'] for exemplo in exemplos]

  batch = processor(text=textos, images=image_inputs, padding=True, return_tensors='pt') #to(device)?

  labels = batch['input_ids'].clone()
  labels[labels == processor.tokenizer.pad_token_id] = -100
  batch['labels'] = batch['input_ids']

  return batch
```

```python
collate_fn(df_train[0:2])
```

```python
trainer = SFTTrainer(
    model=model,
    args =training_args,
    train_dataset=df_train[:100],
    eval_dataset=df_eval[:20],
    peft_config=peft_config,
    data_collator=collate_fn,
    processing_class=processor.tokenizer,
)
```

```python
avaliacao_inicial = trainer.evaluate()
print('avaliação_inicial:', avaliacao_inicial)
print('--'*30)
trainer.train()
```

```python
trainer.save_model(training_args.output_dir)
```

```python
import gc
import time

# https://huggingface.co/learn/cookbook/en/fine_tuning_vlm_trl
def clear_memory():
    if "inputs" in globals():
        del globals()["inputs"]
    if "model" in globals():
        del globals()["model"]
    if "processor" in globals():
        del globals()["processor"]
    if "trainer" in globals():
        del globals()["trainer"]
    if "peft_model" in globals():
        del globals()["peft_model"]
    if "bnb_config" in globals():
        del globals()["bnb_config"]
    time.sleep(2)
    torch.cuda.empty_cache()
    gc.collect()
    time.sleep(2)
    torch.cuda.empty_cache()
    torch.cuda.synchronize()
    time.sleep(2)
    gc.collect()
    time.sleep(2)

    print(f"GPU allocated memory: {torch.cuda.memory_allocated() / 1024**3:.2f} GB")
    print(f"GPU reserved memory: {torch.cuda.memory_reserved() / 1024**3:.2f} GB")


clear_memory()
```
